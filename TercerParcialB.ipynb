For each point o in S do:
Pass a line L parallel to the x-axis through o.
Replace every point in S below L, with its reflection. (For example if L is the x axis, (a,-x) for x>0 will become (a,x) after the reflection). Let the new set of points be S'
The angle of each point p in S', is the right angle of the segment po with the line L. Let us sort the points S' by their angles.
Walk through the sorted points in S'. If there are two consecutive points which are collinear with o - return true.

el siguiente algoritmo se obtuvo de = http://pythonshort.blogspot.com.co/2015/03/intersection-of-two-line-segments-in.html


```python
class Segments:
    
    def intersection(self, s1, s2):
       self.segment_endpoints = []
       left = max(min(s1[0], s1[2]), min(s2[0], s2[2]))
       right = min(max(s1[0], s1[2]), max(s2[0], s2[2]))
       top = max(min(s1[1], s1[3]), min(s2[1], s2[3]))
       bottom = min(max(s1[1], s1[3]), max(s2[1], s2[3]))
        
       if top > bottom or left > right:
          self.segment_endpoints = []
          print 'NO INTERSECTION'
          print self.segment_endpoints
        
       elif top == bottom and left == right:
          self.segment_endpoints.append(left)
          self.segment_endpoints.append(top)
          print 'POINT INTERSECTION'
          print self.segment_endpoints
    
       else:	
          self.segment_endpoints.append(left)
          self.segment_endpoints.append(bottom)
          self.segment_endpoints.append(right)
          self.segment_endpoints.append(top)         
          print 'SEGMENT INTERSECTION'
          print self.segment_endpoints

class Test:
    
    def __init__(self):
       self.solution = Segments()

    def test_0(self):
       s1 = [0, 0, 0, 1]
       s2 = [1, 0, 1, 1]
        
       expected = 'NO INTERSECTION'
       print '------------------------------'
       print 'test_0 ' 
       print 's1 ' + str(s1)
       print 's2 ' + str(s2)
       print 'expected result is ' + expected
       print 'obtained '
       self.solution.intersection(s1, s2)
       
    def test_1(self):
       s1 = [0, 0, 0, 1]
       s2 = [0, 1, 0, 2]
        
       expected = 'POINT INTERSECTION'
       print '------------------------------'
       print 'test_1 '
       print 's1 ' + str(s1)
       print 's2 ' + str(s2)
       print 'expected result is ' + expected
       print 'obtained '
       self.solution.intersection(s1, s2)
       

    def test_2(self):
      s1 = [0, -1, 0, 1]
      s2 = [-1, 0, 1, 0]

      expected = 'POINT INTERSECTION'
      print '-------------------------------'
      print 'test_2 '
      print 's1 ' + str(s1)
      print 's2 ' + str(s2)
      print 'expected result is ' + expected
      print 'obtained '
      self.solution.intersection(s1, s2)
      
    def test_3(self):
      s1 = [0, 0, 2, 0]
      s2 = [1, 0, 10, 0]

      expected = 'SEGMENT INTERSECTION'
      print '-------------------------------'
      print 'test_3 '
      print 's1 ' + str(s1)
      print 's2 ' + str(s2)
      print 'expected result is ' + expected
      print 'obtained '
      self.solution.intersection(s1, s2)
      
   
    def test_4(self):
      s1 = [5, 5, 5, 5]
      s2 = [6, 6, 6, 6]

      expected = 'NO INTERSECTION'
      print '-------------------------------'
      print 'test_4 '
      print 's1 ' + str(s1)
      print 's2 ' + str(s2)
      print 'expected result is ' + expected
      print 'obtained '
      self.solution.intersection(s1, s2)
      
    def test_5(self):
      print '-------------------------------'
      s1 = [10, 0, -10, 0]
      s2 = [5, 0, -5, 0]

      expected = 'SEGMENT INTERSECTION'
      print 'test_5 '
      print 's1 ' + str(s1)
      print 's2 ' + str(s2)
      print 'expected result is ' + expected
      print 'obtained '
      self.solution.intersection(s1, s2)

    def run_tests(self):
       self.test_0()
       self.test_1()
       self.test_2()
       self.test_3()
       self.test_4()
       self.test_5()

test = Test()
test.run_tests()
```
